```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#data = pd.read_csv("crime.csv", index_col="Neighborhood", low_memory=False)
#count = data.value_counts()
#count.plot(kind='bar')
#plt.show()

data = pd.read_csv("crime.csv", low_memory=False)
name = data['Neighborhood']
count = name.value_counts()
count.plot(kind='bar', figsize=(15, 8))
plt.ylabel('Number of Crimes')
plt.show()
```

```python
data['Neighborhood'].unique()
len(data['Neighborhood'].unique())
```

```python
name = data['Neighborhood']
count = name.value_counts()
count = count.nsmallest(n = 10)
count.plot(kind='bar', figsize=(15, 8))
plt.ylabel('Number of Crimes')
plt.show()
```

```python
name = data['NIBRS_Offense_Type']
count = name.value_counts()
count = count.nlargest(n = 10)
count.plot(kind='bar', figsize=(15, 8))
plt.ylabel('Number of Crimes')
plt.show()
```

```python
name = data['Neighborhood']
count = name.value_counts()
count = count.nsmallest(n = 10).index.tolist()
for neighborhood in count:
    sub = data[data['Neighborhood'] == neighborhood]
    crime = sub['NIBRS_Offense_Type'].value_counts().nlargest(n = 5)
    plt.figure(figsize=(6, 6))
    plt.pie(crime, labels=crime.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Crime Distribution in {neighborhood}')
    plt.tight_layout()
    plt.show()
    
```

```python
name = data['Neighborhood']
count = name.value_counts()
count = count.nsmallest(n = 10).index.tolist()
def show_counts(pct, all_vals):
    total = sum(all_vals)
    c = int(round(pct * total / 100.0))
    return f'{c}'
for neighborhood in count:
    sub = data[data['Neighborhood'] == neighborhood]
    crime = sub['NIBRS_Offense_Type'].value_counts().nlargest(n = 5)
    plt.figure(figsize=(6, 6))
    plt.pie(crime, labels=crime.index, autopct=lambda pct: show_counts(pct, crime), startangle=140)
    plt.title(f'Crime Distribution in {neighborhood}')
    plt.tight_layout()
    plt.show()
    
```
