```python
ne_Name = fire['neighborhood']
ne_Name = fire['neighborhood']
neg = ne_Name.value_counts().nsmallest(n=10).index.tolist()
def show_counts(pct, all_vals):
    total = sum(all_vals)
    count = int(round(pct * total / 100.0))
    return f'{count}'
for fire1 in neg:
    sub = fire[fire['neighborhood'] == fire1]
    flame = sub['incident_type'].value_counts()
    plt.figure(figsize=(6, 6))
    plt.pie(
        flame,
        labels=flame.index,
        autopct=lambda pct: show_counts(pct, flame),
        startangle=140
    )
    plt.title(f'Fire Incident Type in {fire1}')
    plt.tight_layout()
    plt.show()
    
```
