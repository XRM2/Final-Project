```python
import pandas as pd
import matplotlib.pyplot as plt
fire= pd.read_csv("Kali'sFiredata.csv")
fire.head(10)
```

Sees how many neigborhoods are in the data set to see if the group has the same data.

```python
unique_order= fire['neighborhood'].unique()
print (len(unique_order))
print (unique_order)
```

**THIS DATA SET IS ALL TOGETHER**

This data is all the negborhoods and it is part of this set so if let's say Carrick scored very low in crime and vechicle crashes but it is the highest in fire when need to be able to see that so it isn't counted to the best neighborhood.

```python
ne_Name = fire['neighborhood']
neg = ne_Name.value_counts()
neg.plot(kind='bar', figsize=(15, 8))
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.show()
```

```python
ne_Name = fire['neighborhood']
neg = (ne_Name.value_counts())
neg=neg.nsmallest(n=10)
negplot = neg.plot(kind='bar', figsize=(15, 8))
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.show()
```

All the neigborhoods that are in the the top 10 for least amount of fires are in running for being the safest negiborhoods
